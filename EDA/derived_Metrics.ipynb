```python
class Solution(object):
    def searchRange(self, nums, target):
        notFound = [-1,-1]
        index = []
        for i in range(len(nums)):
            if nums[i] == target :
                index.append(i)
                break
        for i in range(len(nums) - 1, -1, -1):
            if nums[i] == target:
                index.append(i)
                break

        if target not in nums:
            return notFound

        return index
        
a = Solution()
a.searchRange([5,7,7,8,8,7,10], 6)
name = 'krishna'
print(name[::-1])

```

```python
class Solution(object):
    def isPalindrome(self, x):
        if x<0:
            return False
        original = x
        reversed_num = 0

        while x > 0:
            digit = x % 10           # get last digit
            reversed_num = reversed_num * 10 + digit
            x //= 10                 # remove last digit

        return original == reversed_num
    
x = -121
a1 = Solution()
print(a1.isPalindrome(x))
        
```
